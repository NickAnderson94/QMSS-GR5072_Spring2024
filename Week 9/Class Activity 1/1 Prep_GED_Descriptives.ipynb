```python
import numpy as np
import pandas as pd
```

```python
# import data and prep for analysis
path = r"C:\Users\nicho\Desktop\1 Modern Data Structures - GR5072\QMSS-GR5072_Spring2024\Week 9\Class Activity 1"
df = pd.read_csv(path + r"\2 ged_data.csv")
df["income"] = df["F3ERN2011"]
df["overall"] = 1
df["male"] = np.where(df["female"] == 1, 0, 1)
df["Non-White"] = np.where(df["White"] == 1, 0, 1)
df["dropout"] = np.where(df["high_school_grad"] == 1, 0, np.where(df["ged"] == 1, 0, 1))
df["hours_worked"] = df["F3C02"]
```

```python
def descriptive_stats(df=df, cut_var="female", cols = ["income", "female", "male"]):
    """Function which returns sum and mean of select variables in a df"""
    
    df0 = df[ df[cut_var] == 1 ]
    
    list_sum  = []
    list_mean = []
    cols_sum   = []
    cols_mean  = []

    for col in cols:
        list_sum.append(df0[col].sum())
        list_mean.append(df0[col].mean())
        cols_sum.append(col + "_sum")
        cols_mean.append(col + "_mean")

    sums = np.transpose(pd.DataFrame(list_sum, index = cols_sum))
    means = np.transpose(pd.DataFrame(list_mean, index = cols_mean))
        
    return pd.concat([sums, means], axis=1)
```

```python
col_names = ["income", "female", "male", "moth_ed_1", "moth_ed_2", "moth_ed_3", "moth_ed_4", 
             "moth_ed_5", "moth_ed_6", "moth_ed_7", "moth_ed_8", "region_1", "region_2",
             "region_3", "region_4", "high_school_grad", "ged", "overall", "dropout", 
             "BYTXCSTD", "BYSES1", "hours_worked",
             "Other", "Asian", "Black", "Hispanic", "White", "Non-White", "post_sec_edu"]

overall = descriptive_stats(df, "overall", cols = col_names)
female  = descriptive_stats(df, "female", cols = col_names)
male    = descriptive_stats(df, "male", cols = col_names)
white   = descriptive_stats(df, "White", cols = col_names)
black   = descriptive_stats(df, "Non-White", cols = col_names)
```
